```python
# Load the Library
from autogluon.tabular import TabularDataset,TabularPredictor
```

```python
# The columns in the Dataset
df.columns
```

```python
# if the machine doesn't run the large Dataset create a small sample dataset
small_data = df.sample(100000,random_state=42)
```

## Train and Test Split

```python
# train and test split
train_data = small_data.sample(80000,random_state=42)
```

```python
# train and test split
test_data = small_data.drop(train_data.index)
```

```python
# Create a folder to save the models
save_path = 'uber_predictions'
```

```python
# Prediction
predictor = TabularPredictor(label='fare_amount',path=save_path)
```

## Training using AutoGluon

```python
# fit the prediction onto the dataset
predictor.fit(train_data)
```

```python
# generate the report
predictor.fit_summary()
```

## Validation

```python
# Create a model path
model_path = 'uber_predictions'
```

```python
# load the model
new_predictor = TabularPredictor.load(model_path)
```

```python
# get the best model
new_predictor.model_best
```

```python
# get the y_true values
y_test = test_data['fare_amount']
```

```python
# get the features that is the x_test
test_features = test_data.drop('fare_amount',axis=1)
```

```python
# get the y_pred value
y_pred = new_predictor.predict(test_features)
```

## Evaluation

```python
# get the metrics performance
metrics = predictor.evaluate_predictions(y_true=y_test,y_pred=y_pred,auxiliary_metrics=True)
```

# Advanced AutoGluon

## Presets

```python
# length of the dataset
len(df)
```

```python
# make a small sample dataset
df = df.sample(20000)
```

```python
# find the length of the small data
len(df)
```

```python
# train and test split
train_data = df.sample(19000)
```

```python
# train and test split
test_data = df.drop(train_data.index)
```

```python
# create a folder
save_path = 'my_new_model'
```

```python
# give a preset that is a string code
# medium_quality is the model is not that good at predicting it is of medium quality
# that is not good at predicting but doing things like optimizing for inference time instead
presets = ['medium_quality']
```

```python
# Create a predictor
predictor = TabularPredictor(label='fare_amount',path=save_path)
```

```python
# fit the model using the predictor
# exclude_model_types are optional that is to say not to run the model in the list
# The model that are excluded are neural network model and fastai neural network model
predictor.fit(train_data,presets=presets,excluded_model_types=['NN_TORCH','FASTAI'])
```

```python
# Check the leaderboard
predictor.leaderboard()
```

```python
# create a new predictor using load
new_predictor = TabularPredictor.load(save_path)
```

```python
# Check the leaderboard
new_predictor.leaderboard()
```

## Model Deployment

The **.refit_full()** method in sklearn is used to refit the model on the entire dataset.

```python
# train_data ---> train and validation split
# test_data ---> test split or holdout test data
predictor.refit_full()
```

The **.clone_for_deployment()** method in sklearn creates a new model instance with the same configuration as the original model, but without any pre-trained weights or state.

This is useful when you want to deploy a trained model to a production environment without including any sensitive information.

```python
# clone for deployment
predictor.clone_for_deployment('final_model_for_deployment')
```
